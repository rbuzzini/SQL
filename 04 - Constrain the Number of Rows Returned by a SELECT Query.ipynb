# Constrain the Number of Rows Returned by a SELECT Query

Often it is not convenient to show all rows on screen as you have to scroll. We can use use LIMIT clause to constrain the number of rows returned by a query. This can be used for testing or to prevent consuming excessive resources if the query returns more information than expected.

The following illustrates the syntax of the LIMIT clause:

- SELECT column_list FROM table_name LIMIT row_count;

```python
import mysql.connector as sql
import pandas as pd
```

### 1. Retrieving the first 10 rows of the table of country table

```python
connection = sql.connect(
    host="localhost",
    user="root",
    password="12345"
)
cursor = connection.cursor()
```

```python
pd.read_sql_query("""
    SELECT *
    FROM world.country
    LIMIT 10""",
    connection)
```
